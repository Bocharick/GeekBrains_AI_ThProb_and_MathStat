**3) Непрерывная случайная величина C распределена нормально и задана плотностью распределения  
f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-(x+2)^2) / 32).  
Найдите:  
а). M(X)  
б). D(X)  
в). std(X) (среднее квадратичное отклонение)**

$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}}$$

$$f(x) = \frac{1}{4\sqrt{2 \pi}} e ^ {- \frac{(x + 2)^2}{32}}$$

Исходя из формулы, и зная, что:  
$$a = M(X)$$  
$$\sigma^2 = D(X)$$  

Получаем, что: 
$$M(X) = -2$$  
$$D(X) = 4^2 = 32/2 = 16$$

Следовательно:  
$$std(X) = \sqrt{D(X)}= 4$$

**4) Рост взрослого населения города X имеет нормальное распределение.  
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.  
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:  
а). больше 182 см  
б). больше 190 см  
в). от 166 см до 190 см  
г). от 166 см до 182 см  
д). от 158 см до 190 см  
е). не выше 150 см или не ниже 190 см  
ё). не выше 150 см или не ниже 198 см  
ж). ниже 166 см.**

```python
M = 174
X_std = 8
```

```python
print("M - sigma = ", M - X_std)
print("M - 2*sigma = ", M - 2*X_std)
print("M - 3*sigma = ", M - 3*X_std)
print("M + sigma = ", M + X_std)
print("M + 2*sigma = ", M + 2*X_std)
print("M + 3*sigma = ", M + 3*X_std)
```

```python
# Вероятность больше/меньше 3*sigma:
above_3_sigma_prob = (100 - 99.73)/2
below_3_sigma_prob = (100 - 99.73)/2
```

**Ответы посчитаны в процентах:**

```python
# а). больше 182 см:
# [M+sigma; inf]
13.59 + 2.14 + above_3_sigma_prob
```

```python
# б). больше 190 см
# [M+2*sigma; inf]
2.14 + above_3_sigma_prob
```

```python
# в). от 166 см до 190 см
# [M-sigma; M+2*sigma]
34.13 + 34.13 + 13.59
```

```python
# г). от 166 см до 182 см
# [M-sigma; M+sigma]
34.13 + 34.13
```

```python
# д). от 158 см до 190 см
# [M-2*sigma; M+s*sigma]
13.59 + 34.13 + 34.13 + 13.59
```

```python
# е). не выше 150 см или не ниже 190 см
# 100 - [M-3*sigma; M+2*sigma]
100 - (2.14 + 13.59 + 34.13 + 34.13 + 13.59)
```

```python
# ё). не выше 150 см или не ниже 198 см
# 100 - [M-3*sigma; M+3*sigma]
100 - (2.14 + 13.59 + 34.13 + 34.13 + 13.59 + 2.14)
```

```python
# ж). ниже 166 см.
# [-inf; M-sigma]
below_3_sigma_prob + 2.14 + 13.59
```

**5) На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от 
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?**

```python
M_X = 178
D_X = 25
X_std = np.sqrt(D_X)
X_std
```

```python
np.abs(190 - M_X)/X_std
```
